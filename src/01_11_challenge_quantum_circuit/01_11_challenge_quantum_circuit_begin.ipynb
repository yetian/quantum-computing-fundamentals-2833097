```python
from qiskit import *
from qiskit.visualization import plot_histogram, array_to_latex, plot_bloch_multivector
%matplotlib inline
```

```python
# a quantum circuit with one qubit
qr = QuantumRegister(1)

# a classical register with one bit
cr = ClassicalRegister(1)

# make a quantum circuit with a qubit and a bit
circuit = QuantumCircuit(qr, cr)

# measure the qubit and store the result in the bit
circuit.measure(qr, cr)
```

```python
from qiskit.providers.basic_provider import BasicProvider

# get the simulator
simulator = BasicProvider().get_backend('basic_simulator')

# run the circuit on the simulator
compiled_circuit = transpile(circuit, simulator)

# get the result of the simulation
result = simulator.run(compiled_circuit, shots=1000).result()

# plot the result
plot_histogram(result.get_counts())
```

```python
# remove the measurement
circuit.remove_final_measurements()  # no measurements allowed

from qiskit.quantum_info import Statevector

# get the statevector
statevector = Statevector(circuit)

# print the statevector
array_to_latex(statevector)
```

```python
plot_bloch_multivector(statevector)
```
